```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="ProsusAI/finbert")
```

```python
print(pipe("telsa stock dropped 25%, but meta, google, microsoft railed 30% than yesterday."))
print(pipe("meta reported strong financial results for Q1 2024"))
print(pipe("tesla released its financial results for Q1 2024, missing wall street expectations"))
print(pipe("meta reported strong financial results for Q1 2024, but tesla released its financial results for Q1 2024, missing wall street expectations"))
```
