[ProsusAI_finbert](https://huggingface.co/ProsusAI/finbert)
- 금융 뉴스 데이터셋을 가지고 감정 분석 모델

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="ProsusAI/finbert")
```

```python
# 문장 단위로 모델이 긍정과 부정을 평가

print(pipe("telsa stock dropped 25%, but meta, google, microsoft railed 30% than yesterday.") , "\n")
print(pipe("meta reported strong financial results for Q1 2024"), "\n")
print(pipe("tesla released its financial results for Q1 2024, missing wall street expectations"), "\n")
print(pipe("meta reported strong financial results for Q1 2024, but tesla released its financial results for Q1 2024, missing wall street expectations"), "\n")
print(pipe("삼성전자 주식이 90% 이상 올랐어"), "\n")
```
